# *Rank One Approximation as a Strategy for Wordle*

*Tratando-se das estratégias matemáticas para estimar a palavra secreta, este notebook é relativo a* [*Rank One Approximation as a Strategy for Wordle*](https://arxiv.org/abs/2204.06324) (Aproximação de rank um como estratégia para o termo).

*O artigo propõe para esboçar as melhores estimativas da palavra secreta,  usar a combinação de Rank One Approximation(**Aproximação de rank um**) e latent semantic indexing(**indexação semântica latente**)* para selecionar a palavra que melhor representa o conjunto de possíveis soluções.

### Rank One Approximation(Aproximação de rank um)

Primeiro é necessário entender como a **aproximação de rank baixo**(low rank approximation) é feita. Esta ocorre quando aplicando (single value decomposition) **decomposição em valores singulares** a uma matriz $A_{m×n}$(existem outros métodos). Esta decomposição permite reescrever $A$ como a soma de matrizes de rank um ou posto um. A decomposição em valores singulares é uma fatoração de $A$ na seguinte forma: <p align = "center">
$A=USV^T$ </p>



$U := m×m$ matriz cujas <b>colunas</b> são os vetores singulares $u_i$<br>
$S := m×n$ matriz cujas <b>entradas diagonais</b> são os valores singulares $\sigma_i$<br>
$V := n×n$ matriz cujas <b>linhas</b> são os vetores singulares $v_i$<br></p>

onde:<p align = "center">
$u_i :=$ os autovetores de $AA^T$<br>
$\sigma_i :=$ os autovalores de $AA^T$ (ou $A^TA$)<br>
$v_i :=$ os autovetores de $A^TA$ 

Os autovetores  $U,V$ são organizados em ordem decrescente de acordo com seu autovalor correspondente. Se $m=n$ quer dizer o seguinte:<p align = "center">
$A =\begin{bmatrix}
    u_{11}       & \dots & u_{1m} \\
    \vdots       &       & \vdots \\
    u_{m1}       & \dots & u_{mm}
\end{bmatrix}
\begin{bmatrix}
    \sigma_{1} & 0 & \dots & 0 \\
         0      & \sigma_{2} & \dots & 0 \\
    \vdots      & \vdots &    & \vdots \\
    0      & 0 & \dots & \sigma_{m}
\end{bmatrix}
\begin{bmatrix}
    v_{11}       & \dots & v_{1n} \\
    \vdots       &       & \vdots \\
    v_{n1}       & \dots & v_{nn}
\end{bmatrix}$

e:<p align ="center">
$\sigma_1>\sigma_2...>\sigma_m$<p>

*Agora, como $A$ pode ser escrita como a soma de matrizes de rank um exatamente?*

Ora, observe que a matriz $S$ pode ser reescrita da seguinte forma:
<p align ="center">
$S = S_1 + S_2 + ...+ S_p$<p>

Tal que $p = min(m,n)$. Em nosso exemplo, $p=m=n$. Logo:

<p align ="center">
$S = \begin{bmatrix}
    \sigma_{1} & 0 & \dots & 0 \\
         0      & 0 & \dots & 0 \\
    \vdots      & \vdots &    & \vdots \\
    0      & 0 & \dots & 0
\end{bmatrix}+\begin{bmatrix}
    0 & 0 & \dots & 0 \\
         0      & \sigma_{2} & \dots & 0 \\
    \vdots      & \vdots &    & \vdots \\
    0      & 0 & \dots & 0
\end{bmatrix}+....+\begin{bmatrix}
    0 & 0 & \dots & 0 \\
         0      & 0 & \dots & 0 \\
    \vdots      & \vdots &    & \vdots \\
    0      & 0 & \dots & \sigma_{m}
\end{bmatrix}$<p>

Por conseguinte, temos:

<p align = "center">
$A=USV^T$
 </p>
<p align = "center">
$A = U(S_1+S_2+...+S_p)V^T$<p>
<p align = "center">
$A = \sigma_{1}u_1v_{1}^T + \sigma_{2}u_2v_{2}^T+...+\sigma_{p}u_pv_{p}^T$
<p>




Observe que:<p align = "center">
$u_i = \begin{bmatrix}
u_{1i} \\
u_{2i} \\
\vdots \\
u_{mi}
\end{bmatrix}$ e $ v_i^T=\begin{bmatrix}
v_{i1} & v_{i2} & ... & v_{in} 
\end{bmatrix}$  e $\sigma_i$ correspondente<br><br>

Nos dão uma matris de rank um.
Em geral A não tem rank ou posto completo, contudo a partir daqui usaremos dois teoremas.

* **Teorema 1:** Se A tem rank\posto $r$, então A tem exatamente $r$ valores singulares não negativos.

* **Teorema 2:** Para qualquer $v$ tal que $0\leq v \leq r$ definida uma aproximação de A, $A_v$ onde 
<p align = "center">
$A_v = \sum_{i=1}^{v} \sigma_{i}u_iv_{i}^T$
<p>
Se $v = p = min(m,n)$, define $\sigma_{v+1}=0$. Então

<p align = "center">
$||A-A_v||_{2}= \underset{rank(B) \leq v}{\underset{B \in \mathbb{C}^{m×n}}{\inf}}||A - B||_{2}=\sigma_{v+1}$
<p>

Onde $A_v$ também satisfaz

<p align = "center">
$||A-A_v||_{F}= \underset{rank(B) \leq v}{\underset{B \in \mathbb{C}^{m×n}}{\inf}}||A - B||_{F}= \sqrt{\sigma_{v+1}^2+...+\sigma_{r}^2}$
<p>

A partir destes é possivel concluir a melhor aproximação possivel de $A$ usando apenas dois vetores e um valor singular é:<p align = "center">
$A ≈u_1\sigma_1 v_1^T$</p>

Mais ainda, $u_1$ é o vetor mais significativo ou que mais representa $A$. Basta observar que $||A||_{2} = \sigma_{1}$. Por isso, os vetores que nos são de interesse, são os mais proximos de $u_1$. Usaremos Latent semantic indexing para encontramos quem são.

### Latent Semantic Indexing(indexação semântica latente)

Latent semantic indexing(indexação semântica latente) é usada em mecanismos de pesquisa para encontrar resultados que correspondam perto a um termo pesquisado convertido a um vetor de consulta. É uma maneira sistemática de relacionar o vetor do termo pesquisado com o restante do espaço formado pelo conjunto a ser vasculhado. Baseando-se na proximidade de ângulo θ entre eles e o vetor de consulta, estabelecendo um ranking de similaridade por esta métrica.Usaremos como vetor de busca $u_1$ o auto vetor dominante de $AA^T$. Onde θ é dado por:

<p align="center">
$θ_i = cos^{-1}(\frac{u_1^Tw_i}{|u_1||w_i|})$
</p>

$w_i$ corresponde a um vetor do conjunto a ser vasculhado, uma coluna de $A$. Assim teremos os ângulos de cada $w_i$ em relação a $u_1$ de $AA^T$, tal que $A$ é a matriz que armazena o conjunto de todos os vetores. Quanto menor o ângulo de $w_i$ com $u_i$ melhor $w_i$ representa a totalidade dos vetores em $A$.Pois quanto mais θ é próximo de zero mais na direção de $u_i$ $w_i$ está.

```python
#fUNÇÃO PRA EXTRAIR AS PALAVRAS E LIMPAR OS DADOS


class GetWords:
  
  def __init__(self):
    self.all_words = []
    self._word = ""
    
  def get_letters(self,letter):


    if(ord(letter) <= 64  or  ord(letter) == ord('[') or ord(letter) == ord(']')):
      if(self._word != ''):
        if(len(self._word) == 5):
          self.all_words.append(self._word)
      self._word = ""
      return


    if(ord(letter) >=92 and ord(letter) <= 122):
        self._word += letter 
        return
      

    elif(ord(letter) == ord('á') or ord(letter) == ord('à') or ord(letter) == ord('â') or ord(letter) == ord('ã')):
      self._word += letter #chr(97) 
    elif(ord(letter) == ord('í') or ord(letter) == ord('ì') or ord(letter) == ord('î')):
        self._word += letter#chr(105)  
    elif(ord(letter) == ord('ú') or ord(letter) == ord('ù') or ord(letter) == ord('û') or ord(letter)):
        self._word += letter#chr(117)
    elif(ord(letter) == ord('é') or ord(letter) == ord('è') or ord(letter) == ord('ê') or ord(letter)):
        self._word += letter#chr(101)
    elif(ord(letter) == ord('ó') or ord(letter) == ord('ò') or ord(letter) == ord('ô') or ord(letter) == ord('õ')):
        self._word += letter#chr(111)
    elif(ord(letter) == ord('ç')):
        self._word += letter#chr(99)
        
  
   

   
```

```python

def word_to_a_vector_of_matrix(word):

  vector = [0]*45
  word_matriz = [None]*5
  index = 0
  for i in  word:

    if((ord(i)-97) <= 25):
      vector[ord(i)-97] = 1    
    elif(((ord(i)-224) <= 3) and (ord(i)-224) >= 0): #
      vector[(ord(i)-224)+26] = 1
    elif(((ord(i)-236) <= 2) and (ord(i)-236) >= 0):
      vector[(ord(i)-236)+30] = 1
    elif(((ord(i)-249) <= 2) and (ord(i)-249) >= 0):
      vector[(ord(i)-249)+33] = 1
    elif(((ord(i)-232) <= 2) and (ord(i)-232) >= 0):
      vector[(ord(i)-232)+36] = 1
    elif(((ord(i)-242) <= 3) and (ord(i)-242) >= 0):
      vector[(ord(i)-242)+40] = 1
    elif(ord(i) == 231):
      vector[44] = 1
    
    word_matriz[index] = vector.copy()
    vector.clear()
    vector = [0]*45
    index += 1

  matrix_of_vector = np.reshape(word_matriz,(225,1))
  return matrix_of_vector


    

def U_1_vector(A,A_t):
  
  A_t_A = np.dot(A_t,A)
  Sigma, V = np.linalg.eig(A_t_A)
  Sigma = np.sort(Sigma)
  V_t = []
  V_t.append(V[:,0])
  V_t = np.transpose(V_t)
  U_1 =  np.dot((1/Sigma[len(Sigma)-1])*A,V_t)  

  return U_1


def LSI(A,U_1,all_words):

  U_1_t = np.transpose(U_1)
  angles_vector = np.array([])

  for i in range(0,len(A[0])):
    
    angle = np.arccos((np.dot(U_1_t,A[:,i])) / (np.linalg.norm(U_1)*np.linalg.norm(A[:,i])))
    
    if (np.isnan(angle)):
      angle = np.pi/2

    angles_vector = np.append(angles_vector,angle*180/np.pi)

  best_words = np.array(all_words)[angles_vector.argsort()]
  angles_vector = angles_vector[angles_vector.argsort()]
  
  return best_words,angles_vector




```

## Uma primeira aproximação

(c) Empregando a métrica de similaridade descrita no artigo, qual palavra melhor aproxima o espaço de palpites inicial?

```python
# Digite seu código aqui


dic = open('pt_PT.dic',encoding='UTF-8')
getWords = GetWords()

for i in dic.readlines():

  for letter in i:

    if(ord(letter.lower()) <= 64  or  ord(letter.lower()) == ord('[') or ord(letter.lower()) == ord(']')):
      getWords.get_letters(letter.lower())
      break
    
    
    getWords.get_letters(letter.lower())

A = []
for i in getWords.all_words:

  p = word_to_a_vector_of_matrix(i)
  A.append(p)



A_t = np.reshape(A,(3461, 225))
A = np.transpose(A_t)

print(np.shape(A[0]))

U_1 =  U_1_vector(A,A_t)

print(np.shape(U_1))
best_words,angles = LSI(A,U_1,getWords.all_words)

    

angles = np.real(angles)
print("Mehores Palavras ")
print("=============================================")
print("1:",best_words[0],"ângulo %.4f" % angles[0])
print("2:",best_words[1],"ângulo %.4f" % angles[1])
print("3:",best_words[2],"ângulo %.4f" % angles[2])
print("4:",best_words[3],"ângulo %.4f" % angles[3])
print("5:",best_words[4],"ângulo %.4f\n" % angles[4])

```

```python

def converter(string):

  filter =''
  for i in string:
    
    if ord(i) >= 224 and ord(i) < 228:
      filter+= 'a'
          
    elif ord(i) >= 232 and ord(i) < 235:
      filter+= 'e'
    

    elif ord(i) >= 236 and ord(i) < 239:
      filter+= 'i'
      
    
    elif ord(i) >= 242 and ord(i) < 246:
      filter+= 'o'


    elif ord(i) >= 249 and ord(i) < 252:
      filter+= 'u'
    else:
      filter += i

  return filter;
```
